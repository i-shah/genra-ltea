```python
mc5 = mc5[mc5['QSAR_READY_SMILES'] != '-']
mc5.shape
```

```python
mc5_dict = {k:v for k,v in zip(mc5['DTXSID'], mc5['QSAR_READY_SMILES'])}
```

```python
smi = {k:Chem.MolFromSmiles(v) for k,v in mc5_dict.items()}
```

Generation of Torsion Finger Prints 

```python
from rdkit.Chem.rdMolDescriptors import GetHashedTopologicalTorsionFingerprintAsBitVect
```

```python
tts ={k:GetHashedTopologicalTorsionFingerprintAsBitVect(v) for k,v in smi.items()}
```

```python
tts_list={k:list(v) for k,v in tts.items()}
```

```python
tts_df=pd.DataFrame.from_dict(tts_list).T
```

```python
tts_df.head()
```

```python
tts_df.shape
```

```python
tts_df = tts_df.add_prefix('tptr_')
```

```python
tts_df = tts_df.reset_index()
tts_df.rename(columns = {'index':'DTXSID'}, inplace = True)
tts_df = tts_df.set_index('DTXSID')
```

```python
tts_df.head()
```

```python
tts_df.to_csv('ltea_torsionfp_04.24.2020.tsv', sep='\t')
```

Genearation of Morgan Fingeprints 
    radius=3
    bit length = 2048  

```python
from rdkit.Chem.rdMolDescriptors import GetMorganFingerprintAsBitVect
```

```python
mfp={k:GetMorganFingerprintAsBitVect(v, 3, nBits=2048) for k, v in smi.items()}
```

```python
mfp_list={k:list(v) for k,v in mfp.items()}
```

```python
mfp_df=pd.DataFrame.from_dict(mfp_list).T
```

```python
mfp_df.head()
```

```python
mfp_df = mfp_df.add_prefix('mrgn_')
```

```python
mfp_df = mfp_df.reset_index()
mfp_df.rename(columns = {'index':'DTXSID'}, inplace = True)
mfp_df = mfp_df.set_index('DTXSID')
```

```python
mfp_df.shape
```

```python
mfp_df.head()
```

```python
mfp_df.to_csv('ltea_morganfp_04.24.2020.tsv', sep='\t')
```
